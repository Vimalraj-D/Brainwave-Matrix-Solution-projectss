```python
from diffusers import StableDiffusionPipeline
import torch


pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe.to("cuda" if torch.cuda.is_available() else "cpu")

def generate_image(prompt, width=512, height=512):
    return pipe(prompt, height=height, width=width).images[0]


if __name__ == "__main__":
    prompt="A serene forest bathed in golden sunlight, with tall, ancient trees standing proudly, their leaves forming a lush green canopy. A gentle stream winds through the forest, its crystal-clear water reflecting the vibrant hues of wildflowers scattered along its banks. Birds flutter gracefully above, and a soft mist lingers, adding an ethereal touch to the tranquil scene."
    image = generate_image(prompt)
    image.save("output.png")
```
